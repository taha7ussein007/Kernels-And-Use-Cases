Linear Regression On HR DataSet

```python
import pandas as pd 
hr = pd.read_csv ('Datasets/HR_comma_sep.csv')
```

```python
hr
```

```python
X1= hr[['satisfaction_level' ,'last_evaluation'  ]]
```

```python
Y1 = hr['time_spend_company']
```

```python
from __future__ import print_function
import numpy as np
import statsmodels.api as sm


olsmod = sm.OLS(Y1, X1)
olsres = olsmod.fit()
print(olsres.summary())
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
X1 = hr[['satisfaction_level' ,'last_evaluation' ]]
Y1 = hr['time_spend_company']
reg = lr.fit(X1, Y1)
```

```python
hr['Predicted_Time_to_leave'] = reg.predict(X1)
```

```python
hr['Predicted_Time_to_leave'] = None
for i in hr.index : 
    k = reg.predict(X1) 
    hr['Predicted_Time_to_leave'][i] = k[i]
    
    
    
    
```
